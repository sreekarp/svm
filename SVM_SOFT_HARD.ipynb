```python
from sklearn import datasets
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
#taking a lineraly separable data set
X1, Y1 = datasets.make_blobs(n_samples=100, n_features=2, centers=2, random_state=42, cluster_std=1.0)
```

```python
#non linearly separable dataset
X2, Y2 = datasets.make_moons(n_samples=100, noise=0.3, random_state=42)
```

```python
def plot_decision_boundary(X, y, model, title, xlim=None, ylim=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    ax = plt.gca()

    if xlim is None:
        xlim = ax.get_xlim()
    if ylim is None:
        ylim = ax.get_ylim()

    # Create grid to evaluate model
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 500), np.linspace(ylim[0], ylim[1], 500))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary and margins
    ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    plt.title(title)
```

```python
#using soft margin
linear_soft = svm.SVC(kernel='linear', C=1.0)
linear_soft.fit(X1, Y1)
Y_pred = linear_soft.predict(X1)
print(accuracy_score(Y1,Y_pred))

# Plot decision boundary using the function
plot_decision_boundary(X1, Y1, linear_soft, "Soft Margin SVM Decision Boundary")
```

```python
#using soft margin
linear_soft = svm.SVC(kernel='linear', C=1.0)
linear_soft.fit(X2, Y2)

Y_pred = linear_soft.predict(X2)
print(accuracy_score(Y2,Y_pred))
# Plot decision boundary using the function
plot_decision_boundary(X2, Y2, linear_soft, "Soft Margin SVM Decision Boundary")
```

```python
#using soft margin
rbf_soft = svm.SVC(kernel='rbf')
rbf_soft.fit(X2, Y2)

Y_pred = rbf_soft.predict(X2)
print(accuracy_score(Y2,Y_pred))
# Plot decision boundary using the function
plot_decision_boundary(X2, Y2, rbf_soft, "Soft Margin SVM Decision Boundary")
```

```python

#using soft margin
rbf_hard = svm.SVC(kernel='rbf', C=1e6)
rbf_hard.fit(X2, Y2)
Y_pred = rbf_hard.predict(X2)
print(accuracy_score(Y2,Y_pred))

# Plot decision boundary using the function
plot_decision_boundary(X2, Y2, rbf_hard, "Hard Margin SVM Decision Boundary")
```
